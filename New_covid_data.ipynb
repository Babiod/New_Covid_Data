```python
import pandas as pd
import numpy as np
df=pd.read_csv('Downloads/New_Covid_data.csv')
df
```

```python
df[df['location']=='United States'] 
```

```python
df['DeathPercentage']=df['total_deaths']/df['total_cases']*100
df1= df[['location','date','total_cases','total_deaths','DeathPercentage']]
df1.head()
```

```python
df.dtypes
```

```python
df['PercentagePopulationInfected']=df['total_cases']/df['population']*100
df2= df[['location','total_cases','population','PercentagePopulationInfected']]
df2
```

```python
df['HighestInfectionCount'] = df['total_cases'].max()
df["MaxPercentagePopualtionInfected"] = df['total_cases'].max() /df['population'].max() *100
df3=df[['location','population','HighestInfectionCount','PercentagePopulationInfected']]
df3
df3.to_csv('modified.txt', sep ='t')
df3
```

```python
df4=df.groupby(['location'])['total_deaths'].max()
df4.head()
```

```python
df5=df.groupby(['continent'])['total_deaths'].max()
df5
```

```python
df['sumDeath_Percentage'] = df['new_deaths'].sum()/df['new_cases'].sum() * 100
df.groupby(['date', 'sumDeath_Percentage'])['new_cases'].sum()
```

```python
df['sumDeath_Percentage'] = df['new_deaths'].sum()/df['new_cases'].sum() * 100
df.groupby(['sumDeath_Percentage'])['new_cases'].sum()
```
